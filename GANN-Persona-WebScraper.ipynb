```python
# scrape urls of personas from page
import requests
from bs4 import BeautifulSoup
import pandas as pd

personas_wiki_url = "https://megamitensei.fandom.com/wiki/List_of_Persona_Q2_Personas"
persona_urls = []
root_url = "https://megamitensei.fandom.com"
source = requests.get(personas_wiki_url).text
soup = BeautifulSoup(source, 'lxml')

# scrape tables
for table in soup.find_all('table', {'class': ['customtable pq2', 'table pq2']}): # get the tables
    for tablerow in table.find_all('tr'): # get rows in each table
        persona_url = tablerow.find('td') # get persona url (corresponds to first table data)
        try:
            persona_urls.append(persona_url.a['href'])
        except:
            pass

persona_urls
```

```python
# scrape urls of images of persona's page
import time
import random
import os
import urllib.request

dest_folder = 'personas'
if not os.path.exists(dest_folder):
        os.makedirs(dest_folder)  # create folder if it does not exist

for persona_url in persona_urls:
    # navigate to persona image in website
    source = requests.get(root_url+persona_url+'.jpg').text
    soup = BeautifulSoup(source, 'lxml')
    img_url = soup.find('figure').a['href'] # persona image is the only one to use figure tag

    # download image into folder
    filename = persona_url.split('/')[-1].replace(" ", "_")  # to ensure file name is valid
    filepath = os.path.join(dest_folder, filename)
    urllib.request.urlretrieve(img_url, filepath)
    time.sleep(random.randint(1,10)*0.1) # to prevent bot detection hopefully
    
```
